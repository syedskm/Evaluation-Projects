```python
x = df.drop(['W'],axis=1)
y = df.W
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)

alphas = [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]
best_mae = float('inf')
best_alpha = None
for alpha in alphas:
    ridge_model = Ridge(alpha=alpha)
    ridge_scores = cross_val_score(ridge_model, X_scaled, y, cv=5, scoring='neg_mean_absolute_error')
    ridge_mae = np.mean(-ridge_scores)
    if ridge_mae < best_mae:
        best_mae = ridge_mae
        best_alpha = alpha
        
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
ridge_model = Ridge(alpha=best_alpha)
ridge_model.fit(X_train, y_train)
ridge_predictions = ridge_model.predict(X_test)
ridge_test_mae = mean_absolute_error(y_test, ridge_predictions)
ridge_r2 = r2_score(y_test, ridge_predictions)
print("Ridge Regression Test MAE:", ridge_test_mae)
print("Ridge Regression R-squared:", ridge_r2)
```

```python
lr = LinearRegression()
lr_score = cross_val_score(lr, X_scaled, y, cv=5, scoring='neg_mean_absolute_error')
linear_mae = np.mean(-lr_score)
print("Linear Regression MAE:", linear_mae)

lr.fit(X_train, y_train)
linear_predictions = lr.predict(X_test)
linear_test_mae = mean_absolute_error(y_test, linear_predictions)
linear_r2 = r2_score(y_test, linear_predictions)
print("Linear Regression Test MAE:", linear_test_mae)
print("Linear Regression R-squared:", linear_r2)
```

```python
dt = DecisionTreeRegressor(random_state=42)

dt.fit(X_train, y_train)
tree_predictions = dt.predict(X_test)
tree_test_mae = mean_absolute_error(y_test, tree_predictions)
tree_r2 = r2_score(y_test, tree_predictions)
print("Decision Tree Regression Test MAE:", tree_test_mae)
print("Decision Tree Regression R-squared:", tree_r2)
```

```python
ridge_model = Ridge(alpha=1.0)
ridge_scores = cross_val_score(ridge_model, X_scaled, y, cv=5, scoring='neg_mean_absolute_error')
ridge_mae = np.mean(-ridge_scores)
print("Ridge Regression MAE:", ridge_mae)

# Apply Decision Tree Regression
tree_model = DecisionTreeRegressor(random_state=42)
tree_scores = cross_val_score(tree_model, X_scaled, y, cv=5, scoring='neg_mean_absolute_error')
tree_mae = np.mean(-tree_scores)
print("Decision Tree Regression MAE:", tree_mae)
```
